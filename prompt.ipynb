### Will a Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaraunt near where you are driving. Would you accept that coupon and take a short detour to the restaraunt? Would you accept the coupon but use it on a sunbsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaraunt? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \\$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\\$20 - \\$50). 

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece. 





### Data Description

The attributes of this data set include:
1. User attributes
    -  Gender: male, female
    -  Age: below 21, 21 to 25, 26 to 30, etc.
    -  Marital Status: single, married partner, unmarried partner, or widowed
    -  Number of children: 0, 1, or more than 1
    -  Education: high school, bachelors degree, associates degree, or graduate degree
    -  Occupation: architecture & engineering, business & financial, etc.
    -  Annual income: less than \\$12500, \\$12500 - \\$24999, \\$25000 - \\$37499, etc.
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she buys takeaway food: 0, less than 1, 1 to 3, 4 to 8 or greater
    than 8
    -  Number of times that he/she goes to a coffee house: 0, less than 1, 1 to 3, 4 to 8 or
    greater than 8
    -  Number of times that he/she eats at a restaurant with average expense less than \\$20 per
    person: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    

2. Contextual attributes
    - Driving destination: home, work, or no urgent destination
    - Location of user, coupon and destination: we provide a map to show the geographical
    location of the user, destination, and the venue, and we mark the distance between each
    two places with time of driving. The user can see whether the venue is in the same
    direction as the destination.
    - Weather: sunny, rainy, or snowy
    - Temperature: 30F, 55F, or 80F
    - Time: 10AM, 2PM, or 6PM
    - Passenger: alone, partner, kid(s), or friend(s)


3. Coupon attributes
    - time before it expires: 2 hours or one day

## 6. Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K. 



```python
#Get the drivers who go to bar more than once a month, without kids as passangers 
#and are not widowed
driver_wo_kids=driver_wo_kids[~driver_wo_kids.maritalStatus.isin(['widowed'])]

#Assign group1 for the above group of customers to enable visual distinction in plots
driver_wo_kids['group'] = 'group1'

#Get the drivers who go to bar more than once a month and are under 30
driver_under_30 = bar_coupons[bar_coupons.Bar.isin(["1~3","4~8","gt8"])]
driver_under_30 = driver_under_30[bar_coupons.age.isin(["21","26","below21"])]

#Assign group2 for the above group of customers to enable visual distinction in plots
driver_under_30['group'] = 'group2'

#Get the drivers who go to cheap restaurants more than 4 times a month and income is less than 50K
driver_cheap_rest_under_50k = bar_coupons[bar_coupons.RestaurantLessThan20.isin(["4~8","gt8"])]
driver_cheap_rest_under_50k = driver_cheap_rest_under_50k[driver_cheap_rest_under_50k.income.isin(["$37500 - $49999","$12500 - $24999", "$25000 - $37499", "Less than $12500"])]

#Assign group3 for the above group of customers to enable visual distinction in plots
driver_cheap_rest_under_50k['group'] = 'group3'

combined_df_bar = pd.merge(pd.merge(driver_wo_kids, driver_under_30, how = "outer"), driver_cheap_rest_under_50k, how = "outer")

#Confirming whether the dataset got merged successfully
print(driver_wo_kids.shape)
print(driver_under_30.shape)
print(driver_cheap_rest_under_50k.shape)
print(combined_df_bar.shape)
```

```python
px.histogram(combined_df_bar, x="group", color="couponAcceptanceStatus", title="Comparison of Bar Coupon Acceptance between 3 groups of customers")
```

<h4>Summary:</h4>

<ul>Group1 = Drivers who go to bar more than once a month, do not have kid passangers and were not widowed</ul>

<ul>Group2 = Drivers who go to bar more than once a month and are under 30</ul>

<ul>Group 3 = Drivers who go to cheap restaurants more than 4 times a month and whose income is less than 50K</ul>

<b>Based on the graph above, the drivers in group1 have a more tendency to accept bar coupons than drivers in group 2 and group 3. The drivers in group 3 have more tendency to reject the bar coupons than the other two groups of customers.</b>

### Independent Investigation

Using the bar coupon example as motivation, you are to explore one of the other coupon groups and try to determine the characteristics of passengers who accept the coupons.  

```python
#Determining how likely the frequent bar visitors are going to accept 
#Coffee House coupons compared to the frequent Coffee House visitors

coffee_house_coupons=df_for_analysis.query('coupon=="Coffee House"')
    
#Get the drivers who go to bar more than once a month
freq_bar_visitors=coffee_house_coupons[coffee_house_coupons.Bar.isin(["1~3","4~8","gt8"])]

#Assign "Freq Bar Visitors" as the group for the above group of customers to enable visual distinction in plots
freq_bar_visitors['group'] = 'FreqBarVisitors'

#Get the drivers who go to coffee house more than once a month
freq_coffeehouse_visitors=coffee_house_coupons[coffee_house_coupons.CoffeeHouse.isin(["1~3","4~8","gt8"])]

#Assign "Freq Coffee House Visitors" as the group for the above group of customers to enable visual distinction in plots
freq_coffeehouse_visitors['group'] = 'FreqCoffeeHouseVisitors'

combined_df_coffee = pd.merge(freq_bar_visitors, freq_coffeehouse_visitors, how = "outer")

#Confirming whether the dataset got merged successfully
print(freq_bar_visitors.shape)
print(freq_coffeehouse_visitors.shape)
print(combined_df_coffee.shape)

```

```python
px.histogram(combined_df_coffee, x="group", color="couponAcceptanceStatus", title="Comparison of Coffee House Coupon Acceptance between 2 groups of customers")
```

<h4>Summary:</h4>

<ul>Freq Bar Visitors = Drivers who go to bar more than once a month</ul>

<ul>Group2 = Drivers who go to coffee house more than once a month</ul>

<b>Based on the graph above, the drivers who visit bars frequently have an equal tendency to accept coffee house coupons compared to the frequent coffee house visiting customers.</b>


```python
#Determining how likely the frequent restaurant visitors are going to accept 
#Carry Away coupons

carry_out_coupons=df_for_analysis.query('coupon=="Carry out & Take away"')
    
#Get the drivers who go to less expensive restaurants more than once a month
freq_cheap_rest_visitors=carry_out_coupons[carry_out_coupons.RestaurantLessThan20.isin(["1~3","4~8","gt8"])]

#Assign "FreqInexpRestVisitors" as the group for the above group of customers to enable visual distinction in plots
freq_cheap_rest_visitors['group'] = 'FreqInexpRestVisitors'

#Get the drivers who go to expensive restaurants more than once a month
freq_exp_rest_visitors=carry_out_coupons[carry_out_coupons.Restaurant20To50.isin(["1~3","4~8","gt8"])]

#Assign "FreqExpRestVisitors" as the group for the above group of customers to enable visual distinction in plots
freq_exp_rest_visitors['group'] = 'FreqExpRestVisitors'

combined_df_carryout = pd.merge(freq_cheap_rest_visitors, freq_exp_rest_visitors, how = "outer")

#Confirming whether the dataset got merged successfully
print(freq_cheap_rest_visitors.shape)
print(freq_exp_rest_visitors.shape)
print(combined_df_carryout.shape)
```
